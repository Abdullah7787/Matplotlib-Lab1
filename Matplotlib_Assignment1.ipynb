## Q5: Create three line charts with Legends at left and colors.

```python
data = {
    'Name': ['Ahmad', 'Saleh', 'Ali', 'Faisal'],
    'Age': [25, 30, 22, 28],
    'Major': ['CS ', 'CIS', 'MIS', 'AI']
}

df = pd.DataFrame(data)
plt.plot(df['Name'], df['Age'], marker='o')

plt.xlabel('Name')
plt.ylabel('Age')
plt.title('Age vs. Name')

plt.show()
```

```python
data = {
    'Name': ['Ahmad', 'Saleh', 'Ali', 'Faisal'],
    'Age': [25, 30, 22, 28],
    'Major': ['CS', 'CIS', 'MIS', 'AI'],
    'GPA' : [5,2.1,4.0,3.9]
}

df = pd.DataFrame(data)
fig, ax = plt.subplots()

ax.plot(df['Name'], df['Age'], marker='o', label='Age', color='b')
ax.plot(df['Name'], df['Major'], marker='o', label='Major', color='g')
ax.plot(df['Name'],df['GPA'], marker='o', label='GPA', color='r' )

plt.xlabel('Name')
plt.ylabel('Value')
plt.title('Data Comparison')

ax.legend(loc='upper left')
plt.show()
```

## Q6: Create subplots with three charts and figsize=20,10.

```python
plt.figure(figsize=(20, 10))

plt.subplot(131)
plt.plot(df['Name'], df['Age'], marker='o', label='Age', color='b')
plt.xlabel('Name')
plt.ylabel('Age')



plt.subplot(132)
plt.plot(df['Name'], df['Major'], marker='o', label='Major', color='g')
plt.xlabel('Name')
plt.ylabel('Major')

plt.subplot(133)
plt.plot(df['Name'], df['GPA'], marker='o', label='GPA', color='r')
plt.xlabel('Name')
plt.ylabel('GPA')

plt.subplot(131)
plt.legend(loc='upper left')



plt.show()
```
